# Visualización

Para el área de visualización se han escogido las bibliotecas de ***Matplotlib*** y ***Seaborn***, si bien es posible realizar todos los gráficos de seaborn con matplotlib, con seaborn, los gráficos se encuentran diseñados de forma que sean más agradables y sencillos de usar.

## Las bibliotecas

### **Matplotlib**

[Matplotlib](https://matplotlib.org/) es una biblioteca de gráficos en 2D para Python que permite crear una gran variedad de visualizaciones, desde gráficos sencillos hasta diagramas complejos. Es extremadamente flexible y personalizable, pero puede ser un poco más "baja nivel" en comparación con otras bibliotecas de visualización.

- **Creación de gráficos básicos**: Puedes crear gráficos como líneas, barras, dispersión, histogramas y más.
- **Personalización**: Permite ajustar fácilmente el estilo de los gráficos, el tamaño, el color y otros parámetros visuales.
- **Integración con otras bibliotecas**: Matplotlib se integra bien con otras bibliotecas como NumPy y pandas, lo que lo hace ideal para trabajos con grandes volúmenes de datos numéricos.

**Ejemplo básico**

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Gráfico de Línea")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

### Seaborn
[Seaborn](https://seaborn.pydata.org/) es una biblioteca de visualización basada en Matplotlib que ofrece una interfaz más amigable y atractiva para crear gráficos estadísticos. Seaborn está diseñado para trabajar fácilmente con ***pandas DataFrames*** y para crear gráficos complejos con menos código que Matplotlib.

- **Gráficos estadísticos**: Seaborn facilita la creación de gráficos como distribuciones, diagramas de cajas (boxplots), diagramas de violín, gráficos de dispersión con líneas de regresión, etc.

- **Mejor apariencia**: Viene con una serie de estilos y paletas de colores que mejoran la estética de los gráficos por defecto.
Visualización de relaciones complejas: Es ideal para explorar la relación entre variables de diferentes tipos de datos (por ejemplo, categorías, distribuciones y regresión).

***Ejemplo básico:***

```python
import seaborn as sns

# Datos de ejemplo
data = sns.load_dataset("tips")

# Gráfico de dispersión con regresión
sns.regplot(x="total_bill", y="tip", data=data)
plt.title("Relación entre la cuenta total y la propina")
plt.show()
```

### Diferencias clave:

- **Matplotlib** es más flexible y te permite un control detallado de cada aspecto del gráfico.

- **Seaborn** simplifica la creación de gráficos complejos y ofrece un diseño visual más atractivo por defecto, especialmente para gráficos estadísticos.

En resumen, Matplotlib es una excelente opción cuando se necesita control total sobre los gráficos, mientras que Seaborn facilita la creación rápida de gráficos estadísticos de alta calidad con menos código.

## Hisograma

### ¿Qué es un histograma y cuándo deberíamos usarlo?

Un **histograma** es un gráfico que muestra la distribución de un conjunto de datos, es decir, cómo se distribuyen los valores de una variable numérica a lo largo de un rango de intervalos (llamados bins o "contenedores"). Cada barra del histograma representa la frecuencia (o el número de observaciones) que cae dentro de un intervalo específico. Este tipo de gráfico es especialmente útil para:

- **Visualizar la distribución** de una variable continua (como la edad, el ingreso, la altura, etc.).

- **Detectar la forma de la distribución**: si es simétrica, sesgada a la derecha o a la izquierda, o si tiene varios picos (distribución multimodal).

- **Identificar valores atípicos** o anomalías en los datos.

- **Explorar la densidad de los datos**: qué tan concentrados o dispersos están los valores en diferentes rangos.

Los histogramas son muy útiles en el análisis exploratorio de datos, ya que permiten tener una vista general rápida de la distribución de los datos.

```python
plt.figure(figsize=(10, 6))  # Establece el tamaño de la figura
sns.histplot(df['Age'].dropna(), bins=30, kde=True)  # Crea el histograma con KDE
plt.title("Distribución de Edad de los Pasajeros")  # Título del gráfico
plt.xlabel("Edad")  # Etiqueta del eje X
plt.ylabel("Frecuencia")  # Etiqueta del eje Y
plt.savefig("viz/age_distribution.png")  # Guarda el gráfico en un archivo
plt.show()  # Muestra el gráfico
```

***Explicación paso a paso***:
- `plt.figure(figsize=(10, 6))`:

    Esta línea establece el tamaño del gráfico. En este caso, el gráfico tendrá un ancho de 10 pulgadas y una altura de 6 pulgadas. Esto es útil para asegurarse de que el gráfico tenga suficiente espacio para visualizarse correctamente, especialmente si los datos son densos o el gráfico tiene muchas barras.

- `sns.histplot(df['Age'].dropna(), bins=30, kde=True)`:

    - `sns.histplot` es la función de Seaborn utilizada para crear histogramas.
    - `df['Age']`: Especifica que estamos trabajando con la columna `Age` del DataFrame `df`, que contiene los datos de la edad de los pasajeros.
    - `.dropna()`: Se utiliza para eliminar cualquier valor `NaN` (valores faltantes) que pueda existir en la columna `Age`. Esto es importante, ya que los valores faltantes no pueden ser representados en un histograma.
    - `bins=30`: Define el número de intervalos o "bins" en los que se divide el rango de datos. En este caso, el rango de edades se dividirá en 30 intervalos. El número de bins influye en la precisión de la representación: demasiados bins pueden hacer que el gráfico se vea ruidoso, mientras que pocos bins pueden ocultar detalles importantes de la distribución.
    - `kde=True`: Este parámetro agrega una **Curva de Densidad Kernel (KDE)** al histograma. La KDE es una suavización que permite ver la distribución de los datos de manera más continua, ayudando a identificar patrones, como si los datos siguen una distribución normal o si tienen múltiples picos. Es una herramienta útil para una visualización más clara de la forma de los datos.

- `plt.title("Distribución de Edad de los Pasajeros")`:

    Establece el título del gráfico. En este caso, el gráfico mostrará el título "Distribución de Edad de los Pasajeros", lo que indica claramente qué se está visualizando.

- `plt.xlabel("Edad") y plt.ylabel("Frecuencia")`:

    Estas líneas añaden etiquetas a los ejes X e Y del gráfico. El eje X representa las edades, mientras que el eje Y representa la frecuencia (el número de pasajeros en cada intervalo de edad).

- `plt.savefig("viz/age_distribution.png")`:

    Esta línea guarda el gráfico como un archivo de imagen en formato PNG en una carpeta llamada viz. El gráfico se guardará con el nombre age_distribution.png. Esto es útil si deseas compartir o guardar los resultados visuales sin necesidad de recrear el gráfico cada vez.

- `plt.show()`:

    Finalmente, esta línea muestra el gráfico en pantalla. Esto es necesario para que el gráfico sea visible en el entorno de Jupyter Notebook o en cualquier otro entorno de desarrollo.

## ¿Qué es un Countplot?

Un countplot es un tipo de gráfico de barras que se utiliza para mostrar la frecuencia (o el número de ocurrencias) de categorías en una variable categórica. En lugar de mostrar la distribución de una variable continua como en los histogramas, un countplot muestra cuántas veces aparece cada categoría de una variable en los datos.

***¿Cuándo usarlo?***

- Los **countplots** son útiles cuando queremos ver la distribución de una variable categórica.

- También son útiles para comparar la frecuencia de diferentes categorías entre grupos, especialmente cuando se usan argumentos como hue, que permite visualizar una variable adicional dentro de las barras.

```python
# El código que proporcionas crea un countplot para visualizar cómo se distribuyen
# los géneros (Sex) de los pasajeros y cómo se correlaciona esa distribución con la
# supervivencia (Survived).

plt.figure(figsize=(8, 6))  # Establece el tamaño de la figura
sns.countplot(x='Sex', hue='Survived', data=df)  # Crea el countplot
plt.title("Supervivencia por Género")  # Título del gráfico
plt.xlabel("Género")  # Etiqueta del eje X
plt.ylabel("Número de Pasajeros")  # Etiqueta del eje Y
plt.legend(["No Sobrevivió", "Sobrevivió"])  # Leyenda para los valores de Survived
plt.savefig("viz/survival_by_gender.png")  # Guarda el gráfico en un archivo
plt.show()  # Muestra el gráfico
```

***Explicación paso a paso***:
- `plt.figure(figsize=(8, 6))`:

    Esta línea establece el tamaño del gráfico. En este caso, el gráfico tendrá un ancho de 8 pulgadas y una altura de 6 pulgadas. Esto es útil para asegurarse de que el gráfico tenga suficiente espacio para visualizarse correctamente, especialmente si los datos son densos o el gráfico tiene muchas barras.

- `sns.countplot(x='Sex', hue='Survived', data=df)`:

    - `sns.countplot`: Es la función de Seaborn que crea un gráfico de barras para mostrar la frecuencia de categorías.

    - `x='Sex'`: Esto especifica que la variable en el eje X es `Sex`, que representa el género de los pasajeros (M para hombres y F para mujeres).

    - `hue='Survived'`: El argumento `hue` permite agregar una segunda variable para desglosar la categoría principal. En este caso, estamos usando la variable Survived (que indica si el pasajero sobrevivió o no: 1 = sobrevivió, 0 = no sobrevivió). Esto hará que las barras de cada categoría de género se dividan en dos partes: una para los pasajeros que sobrevivieron y otra para los que no.

    - `data=df`: El DataFrame df es el conjunto de datos que contiene las variables Sex y Survived, así como otras columnas relevantes.

Este gráfico generará barras para cada valor de Sex (es decir, "male" y "female"), y dentro de cada barra habrá una subdivisión en función de si los pasajeros sobrevivieron o no. Es decir, cada barra tendrá dos colores, uno para los que sobrevivieron y otro para los que no.

- `plt.title("Supervivencia por Género")`:

    Establece el título del gráfico. En este caso, el gráfico mostrará el título "Supervivencia por Género", lo que indica claramente qué se está visualizando.

- `plt.xlabel("Género")` y `plt.ylabel("Número de Pasajeros")`:

    - `xlabel("Género")`: Etiqueta el eje X, indicando que las barras representan los géneros de los pasajeros.
    - `ylabel("Número de Pasajeros")`: Etiqueta el eje Y, indicando que las barras muestran el número de pasajeros de cada género.

- `plt.legend(["No Sobrevivió", "Sobrevivió"])`:

    Añade una leyenda al gráfico que explica qué color representa a los pasajeros que sobrevivieron y cuáles no. El primer color en las barras será para los pasajeros que no sobrevivieron, y el segundo color será para los que sobrevivieron.

- `plt.savefig("viz/survival_by_gender.png")`:

    Esta línea guarda el gráfico como un archivo de imagen en formato PNG en una carpeta llamada viz. El gráfico se guardará con el nombre survival_by_gender.png. Esto es útil si deseas compartir o guardar los resultados visuales sin necesidad de recrear el gráfico cada vez.

- `plt.show()`:

    Finalmente, esta línea muestra el gráfico en pantalla. Esto es necesario para que el gráfico sea visible en el entorno de Jupyter Notebook o en cualquier otro entorno de desarrollo.

```python
plt.figure(figsize=(8, 6))  # Establece el tamaño de la figura
sns.countplot(x='Pclass', hue='Survived', data=df)  # Crea el countplot por clase y supervivencia
plt.title("Supervivencia por Clase de Pasajero")  # Título del gráfico
plt.xlabel("Clase")  # Etiqueta del eje X
plt.ylabel("Número de Pasajeros")  # Etiqueta del eje Y
plt.legend(["No Sobrevivió", "Sobrevivió"])  # Leyenda para los valores de Survived
plt.savefig("viz/survival_by_class.png")  # Guarda el gráfico en un archivo
plt.show()  # Muestra el gráfico
```

## Qué es un scatterplot?

Un **scatterplot** es un gráfico que muestra puntos en un plano cartesiano, donde cada punto representa una observación en el dataset. Los valores de las variables se representan en los ejes X y Y, y cada punto es una combinación de estos valores. Este tipo de gráfico es muy útil para:

- ***Explorar la relación entre dos variables***: Puedes ver si existe una tendencia o correlación entre las variables. Por ejemplo, si una variable aumenta cuando la otra lo hace, eso podría indicar una correlación positiva.

- ***Identificar patrones y ***: Un scatterplot permite ver si hay una relación lineal o no lineal entre las variables, si los datos están agrupados en ciertos puntos o si existen valores atípicos (outliers).

- ***Visualizar agrupamientos o segmentos***: Si usamos el argumento hue o alguna otra categoría, el gráfico puede mostrar cómo se agrupan los datos según alguna característica adicional, como la supervivencia o la clase de los pasajeros.

```python
dfc = df.copy()
df['Survived'] = df['Survived'].astype('category')
plt.figure(figsize=(10, 6))  # Establece el tamaño de la figura
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)  # Crea el scatterplot
plt.title("Relación entre Edad y Tarifa")  # Título del gráfico
plt.xlabel("Edad")  # Etiqueta del eje X
plt.ylabel("Tarifa")  # Etiqueta del eje Y
plt.legend(["No Sobrevivió", "Sobrevivió"])  # Leyenda para los valores de Survived
plt.savefig("viz/age_vs_fare.png")  # Guarda el gráfico en un archivo
plt.show()  # Muestra el gráfico
```

## Matriz de Correlación:

Una matriz de correlación es una tabla que muestra las relaciones estadísticas entre un conjunto de variables numéricas. Cada valor en la matriz representa el coeficiente de correlación entre dos variables, que puede estar en un rango de ***-1 a 1***:

- `1`: Indica una correlación perfecta positiva. Cuando una variable aumenta, la otra también lo hace de manera proporcional.

- `-1`: Indica una correlación perfecta negativa. Cuando una variable aumenta, la otra disminuye de manera proporcional.

- `0`: No hay correlación lineal entre las variables.

La ***correlación*** mide cómo varían dos variables de manera conjunta. Si las variables están altamente correlacionadas (ya sea de manera positiva o negativa), esto puede ser un indicativo de que tienen una relación funcional.

### ¿Cuándo usar la matriz de correlación?

Se utiliza cuando quieres explorar relaciones entre variables numéricas, lo cual es útil para:

- ***Detección de relaciones lineales***: La matriz ayuda a identificar qué variables están fuertemente correlacionadas entre sí, lo que podría indicar que una variable es un buen predictor de otra.

- ***Eliminación de variables redundantes***: Si dos variables tienen una alta correlación (por ejemplo, una correlación mayor a 0.8 o 0.9), podrías considerar eliminar una de ellas en modelos predictivos, ya que ambas aportan información similar.

- ***Análisis exploratorio de datos***: Identificar patrones y relaciones entre las variables, lo que puede orientar el análisis posterior.

### ¿Cómo interpretamos una matriz de correlación?

La matriz de correlación puede ayudarnos a identificar las relaciones entre variables:

- ***Correlación positiva***: Si el valor de correlación entre dos variables es cercano a `1`, indica que ambas variables tienden a aumentar juntas. Por ejemplo, en el caso de un dataset como el del Titanic, una correlación alta entre las columnas `SibSp` (número de hermanos/cónyuges) y `Parch` (número de padres/hijos a bordo) podría indicar que las personas que viajan con más familiares tienen un número similar de hermanos y padres a bordo.

- ***Correlación negativa***: Si el valor de correlación es cercano a `-1`, significa que a medida que una variable aumenta, la otra disminuye. Por ejemplo, una posible correlación negativa entre `Age` y `Fare` (con los niños probablemente pagando menos por el billete que los adultos) podría verse en ciertos segmentos de la población.

- ***Correlación cercana a cero***: No hay una relación lineal clara entre las dos variables. Esto puede indicar que las variables no están directamente relacionadas.

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df[['Pclass', 'Survived', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Mapa de Calor de Correlaciones")
plt.savefig("viz/heatmap_correlation.png")
plt.show()
```
