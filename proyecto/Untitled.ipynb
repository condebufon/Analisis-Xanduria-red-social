```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv('datos/Publico2.csv', sep=';')  
```

```python
df
```

```python
# 1. Distribucion por edades
plt.figure(figsize=(10, 6))
sns.histplot(df['Edad'].dropna(), bins=30, kde=True)
plt.title('Distribución de edadcde los seguidores')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.savefig('visualizaciones/distribucion_edad.png')
plt.show()  # Display the plot
plt.close()
```

# Conclucion

Concentración en Grupos Jóvenes: Es probable que el histograma muestre una mayor concentración de seguidores en los grupos más jóvenes, como "18-24" y "25-34", lo que podría indicar que estos segmentos son más activos o representativos dentro del público objetivo.

Disminución en Grupos Mayores: A medida que se avanza hacia grupos etarios más altos, como "55-64" y "65+", es posible que se observe una disminución significativa en la frecuencia, sugiriendo que hay menos seguidores en estas categorías.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame(df)

# Transformar los datos para el countplot
df_melted = df.melt(id_vars='Edad', value_vars=['Hombres', 'Mujeres'], 
                    var_name='Género', value_name='Tasa')

# Eliminar filas con valores NaN
df_melted = df_melted.dropna()

# Crear el gráfico de countplot
plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, x='Edad', y='Tasa', hue='Género', palette='Set2')

# Añadir títulos y etiquetas
plt.title('Comparación de la Tasa de Seguidores entre Hombres y Mujeres')
plt.xlabel('Grupo de Edad')
plt.ylabel('Tasa (%)')
plt.legend(title='Género')

# Mostrar el gráfico
plt.show()
```

# Conclucion
Diferencias en la Tasa de Seguidores: En general, las tasas de seguidores son significativamente más altas para los hombres en comparación con las mujeres en la mayoría de los grupos de edad. Esto sugiere que, al menos en este conjunto de datos, los hombres tienen una mayor presencia o atractivo en la plataforma analizada.

Tendencias por Edad: La gráfica revela que los grupos de edad más jóvenes (18-24 y 25-34) muestran tasas más elevadas tanto para hombres como para mujeres. Sin embargo, la diferencia entre géneros se hace más pronunciada a medida que se avanza hacia grupos de edad mayores, donde la tasa de seguidores femeninos tiende a disminuir considerablemente.

Potencial para el Crecimiento Femenino: A pesar de las tasas más bajas, hay un potencial significativo para el crecimiento en la participación femenina en redes sociales o plataformas similares, especialmente en los grupos de edad más jóvenes. Esto podría indicar una oportunidad para estrategias de marketing o campañas dirigidas a aumentar la visibilidad y el engagement entre las mujeres.
